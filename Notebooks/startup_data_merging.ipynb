```python
import pandas as pd

crunch = pd.read_csv("cleaned_crunchbase_companies.csv")
startup = pd.read_csv("startup data.csv")

crunch.head()
```

```python
crunchbase = crunch[['name','market','funding_total_usd','status','country_code','funding_rounds','founded_year']]
crunchbase.head()
```

```python
crunchbase = crunchbase.rename(columns={"market":"industry"})
crunchbase.head()
```

```python
startup.head()
```

```python
subset_startup = startup[['name','state_code','founded_at','funding_rounds','funding_total_usd','category_code','status']]
subset_startup.head()
```

```python
subset_startup = subset_startup.rename(columns = {'category_code':'industry','founded_at':'founded_year','state_code':'country_code'})
subset_startup.head()
```

```python
subset_startup['country_code'] = 'USA'
subset_startup.head()
```

```python
subset_startup['founded_year'] = pd.to_datetime(subset_startup['founded_year']).dt.year
subset_startup.head()
```

```python
subset_startup.info()
```

```python
crunchbase.info()
```

```python
crunchbase.isnull().sum()
```

```python
crunchbase.dropna(inplace=True)
crunchbase.isnull().sum()
```

```python
crunchbase.info()
```

```python
crunchbase['founded_year'] = crunchbase['founded_year'].astype('Int64').astype(str)
crunchbase.head()
```

```python
crunchbase.info()
```

```python
subset_startup['founded_year'] = subset_startup['founded_year'].astype(str)
subset_startup.info()
```

```python
crunchbase['name'] = crunchbase['name'].str.strip()
subset_startup['name'] = subset_startup['name'].str.strip()
```

```python
crunchbase['name'].dtype
subset_startup['name'].dtype
```

```python
startup_df = pd.concat([crunchbase,subset_startup], ignore_index=True)
startup_df.head()
```

```python
startup_df.info()
```

```python
startup_df.drop_duplicates()
```

```python
countries = startup_df['country_code'].unique()
countries
```

```python
startup_df = startup_df[startup_df['country_code'].str.lower() != 'unknown']
startup_df['country_code'].unique()
```

```python
import pycountry

def convert_country_code(code):
    try:
        return pycountry.countries.get(alpha_3=code).name
    except:
        return code

startup_df['country_code'] = startup_df['country_code'].apply(convert_country_code)
```

```python
startup_df['country_code'].unique()
```

```python
startup_df = startup_df.rename(columns = {'country_code':"country"})
startup_df.head()
```

```python
startup_df['founded_year'].unique()
```

```python
startup_df['founded_year'] = pd.to_numeric(
    startup_df['founded_year'],
    errors = 'coerce'
) 
startup_df = startup_df[startup_df['founded_year'] >= 2000]
startup_df['founded_year'].unique()
```

```python
startup_df.info()
```

```python
startup_df.to_csv('startup_companies.csv', index=False)
```
