```python
"""
Google Trends with Multiple Keywords
"""

from pytrends.request import TrendReq
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import time
from datetime import datetime

print("=" * 80)
print("Multi-Keyword Google Trends Collection")
print("=" * 80)

# Configuration
MIN_YEAR = 2000
MAX_YEAR = 2020

# MULTI-KEYWORD MAPPING (2-5 keywords per industry)


INDUSTRY_KEYWORDS = {
    'E-Commerce & Retail': [
        'ecommerce',
        'online shopping',
        'retail technology',
        'online retail'
    ],
    
    'Technology & Software': [
        'software',
        'saas',
        'cloud computing',
        'enterprise software'
    ],
    
    'Healthcare & Life Sciences': [
        'digital health',
        'telemedicine',
        'healthcare technology',
        'medical devices',
        'biotech'
    ],
    
    'Finance & Fintech': [
        'fintech',
        'digital banking',
        'payment technology',
        'cryptocurrency',
        'blockchain finance'
    ],
    
    'Media & Entertainment': [
        'streaming services',
        'digital media',
        'content creation',
        'entertainment technology'
    ],
    
    'Education & Training': [
        'edtech',
        'online learning',
        'elearning',
        'educational technology'
    ],
    
    'Real Estate & Construction': [
        'proptech',
        'real estate technology',
        'construction technology',
        'smart buildings'
    ],
    
    'Travel & Hospitality': [
        'travel technology',
        'online booking',
        'hotel technology',
        'tourism technology'
    ],
    
    'Food & Beverage': [
        'food delivery',
        'food technology',
        'restaurant technology',
        'meal kit'
    ],
    
    'Transportation & Logistics': [
        'logistics technology',
        'supply chain',
        'delivery services',
        'transportation technology',
        'fleet management'
    ],
    
    'Marketing & Advertising': [
        'digital marketing',
        'martech',
        'advertising technology',
        'marketing automation'
    ],
    
    'Internet & Web Services': [
        'web services',
        'internet technology',
        'web applications',
        'online services'
    ],
    
    'Data & Analytics': [
        'data analytics',
        'big data',
        'business intelligence',
        'data science'
    ],
    
    'Gaming': [
        'video games',
        'gaming',
        'esports',
        'mobile gaming'
    ],
    
    'Social & Community': [
        'social media',
        'social networking',
        'online community',
        'social platform'
    ],
    
    'Consumer & Lifestyle': [
        'consumer technology',
        'lifestyle products',
        'consumer goods',
        'personal technology'
    ],
    
    'Manufacturing & Industrial': [
        'industrial automation',
        'manufacturing technology',
        'industry 4.0',
        'smart manufacturing'
    ],
    
    'Sports & Recreation': [
        'fitness technology',
        'sports technology',
        'wellness',
        'fitness apps'
    ],
    
    'Communications & Telecom': [
        'telecommunications',
        'communication technology',
        '5g technology',
        'telecom'
    ],
    
    'Energy & Clean Tech': [
        'clean energy',
        'renewable energy',
        'cleantech',
        'solar energy',
        'energy technology'
    ],
    
    'Enterprise & Business Services': [
        'business software',
        'enterprise solutions',
        'business services',
        'b2b software'
    ],
    
    'Security & Privacy': [
        'cybersecurity',
        'information security',
        'data privacy',
        'network security'
    ],
    
    'AI & Machine Learning': [
        'artificial intelligence',
        'machine learning',
        'deep learning',
        'ai technology'
    ],
    
    'Customer Service & CRM': [
        'crm software',
        'customer service',
        'customer experience',
        'help desk software'
    ],
    
    'HR & Recruiting': [
        'recruitment technology',
        'hr software',
        'talent management',
        'recruiting'
    ],
    
    'Government & Politics': [
        'govtech',
        'government technology',
        'civic technology',
        'public sector technology'
    ]
}



# Get trend slope with multiple keywords


def get_trend_slope(keywords, year, max_retries=3):
    """
    Get average trend slope across multiple keywords
    
    This approach:
    1. Queries each keyword separately
    2. Calculates slope for each
    3. Returns the AVERAGE slope
    
    Parameters:
    -----------
    keywords : list
        List of search terms (e.g., ['fintech', 'digital banking'])
    year : int
        Year to analyze
    max_retries : int
        Number of retry attempts per keyword
    
    Returns:
    --------
    dict with avg_slope, slopes_per_keyword, success_count
    """
    slopes = []
    keyword_results = {}
    
    for keyword in keywords:
        for attempt in range(max_retries):
            try:
                pytrends = TrendReq(hl='en-US', tz=360, timeout=(10, 25))
                timeframe = f'{year}-01-01 {year}-12-31'
                
                pytrends.build_payload([keyword], timeframe=timeframe)
                data = pytrends.interest_over_time()
                
                if not data.empty and keyword in data.columns:
                    search_values = data[keyword].values
                    
                    if len(search_values) > 1:
                        X = np.arange(len(search_values)).reshape(-1, 1)
                        y = search_values
                        
                        model = LinearRegression()
                        model.fit(X, y)
                        slope = model.coef_[0]
                        
                        slopes.append(slope)
                        keyword_results[keyword] = slope
                        break  # Success, move to next keyword
                    else:
                        keyword_results[keyword] = 0
                        break
                else:
                    keyword_results[keyword] = 0
                    break
                    
            except Exception as e:
                if attempt < max_retries - 1:
                    time.sleep((attempt + 1) * 3)
                else:
                    keyword_results[keyword] = 0
                    break
        
        # Small delay between keywords
        time.sleep(1)
    
    # Calculate average slope
    avg_slope = np.mean(slopes) if slopes else 0
    
    return {
        'avg_slope': avg_slope,
        'keyword_slopes': keyword_results,
        'success_count': len(slopes),
        'total_keywords': len(keywords)
    }






industries = startup['category'].unique().tolist()
years = startup['founded_year'].dropna().unique()
years = sorted([int(y) for y in years if MIN_YEAR <= y <= MAX_YEAR])

print(f"\n Data Scope:")
print(f"   Industries: {len(industries)}")
print(f"   Years: {len(years)} (from {min(years)} to {max(years)})")

# Count total keyword queries
total_keywords = sum(len(INDUSTRY_KEYWORDS.get(ind, [ind])) for ind in industries)
total_queries = total_keywords * len(years)

print(f"   Keywords per industry: 2-5")
print(f"   Total keyword queries: {total_queries}")
print(f"   Estimated time: {(total_queries * 2) / 60:.1f} minutes")



# MAIN COLLECTION LOOP


print("\n" + "=" * 80)
print("Starting Multi-Keyword Data Collection")
print("=" * 80)
print("\n Started at:", datetime.now().strftime("%Y-%m-%d %H:%M:%S"))

results = []
start_time = time.time()
total_combos = len(industries) * len(years)
current_combo = 0

for industry in industries:
    keywords = INDUSTRY_KEYWORDS.get(industry, [industry.lower()])
    
    print(f"\n{'='*80}")
    print(f"Processing: {industry}")
    print(f"   Keywords ({len(keywords)}): {', '.join(keywords)}")
    print(f"{'='*80}")
    
    for year in years:
        current_combo += 1
        elapsed = time.time() - start_time
        avg_time = elapsed / current_combo if current_combo > 0 else 0
        eta_seconds = avg_time * (total_combos - current_combo)
        eta_minutes = eta_seconds / 60
        
        # Progress bar
        progress = (current_combo / total_combos) * 100
        bar_length = 40
        filled = int(bar_length * current_combo / total_combos)
        bar = '' * filled + '' * (bar_length - filled)
        
        print(f"\n  [{bar}] {progress:.1f}%")
        print(f"  Year: {year} | Progress: {current_combo}/{total_combos} | ETA: {eta_minutes:.1f} min")
        print(f"   Querying {len(keywords)} keywords...", end=" ", flush=True)
        
        # Get multi-keyword trend slope
        result = get_trend_slope(keywords, year)
        
        # Store result
        results.append({
            'industry': industry,
            'year': year,
            'trend_slope': result['avg_slope'],
            'num_keywords': result['total_keywords'],
            'successful_keywords': result['success_count'],
            'keyword_details': str(result['keyword_slopes'])
        })
        
        # Print result
        success_pct = (result['success_count'] / result['total_keywords']) * 100
        print(f"Avg Slope: {result['avg_slope']:+.4f} ({result['success_count']}/{result['total_keywords']} keywords, {success_pct:.0f}%)")
        
        # Show individual keyword slopes
        if result['keyword_slopes']:
            for kw, slope in result['keyword_slopes'].items():
                print(f"      {kw}: {slope:+.4f}")
        
        # Rate limiting
        time.sleep(2)


# ============================================================================
# SAVE RESULTS
# ============================================================================

print("\n" + "=" * 80)
print("Saving Results")
print("=" * 80)

trends_df = pd.DataFrame(results)
output_file = 'google_trends.csv'
trends_df.to_csv(output_file, index=False)



# ============================================================================
# SUMMARY STATISTICS
# ============================================================================

print("\n" + "=" * 80)
print("Summary Statistics")
print("=" * 80)

total_time = time.time() - start_time
print(f"\n  Total time: {total_time/60:.1f} minutes")
print(f" Average time per industry-year: {total_time/total_combos:.1f} seconds")

print(f"\n Trend Slope Statistics:")
print(trends_df['trend_slope'].describe())

print(f"\n Keyword Success Rate:")
print(f"   Average keywords per industry: {trends_df['num_keywords'].mean():.1f}")
print(f"   Average successful queries: {trends_df['successful_keywords'].mean():.1f}")
print(f"   Overall success rate: {(trends_df['successful_keywords'].sum() / trends_df['num_keywords'].sum() * 100):.1f}%")

# Top growing industries
print(f"\n Top 5 Growing Industries (Highest Avg Slope):")
top_growing = trends_df.groupby('industry')['trend_slope'].mean().sort_values(ascending=False).head(5)
for i, (industry, slope) in enumerate(top_growing.items(), 1):
    print(f"  {i}. {industry:40s} {slope:+.4f}")

# Most volatile industries
print(f"\n Most Volatile Industries (Highest Slope Variation):")
volatility = trends_df.groupby('industry')['trend_slope'].std().sort_values(ascending=False).head(5)
for i, (industry, std) in enumerate(volatility.items(), 1):
    print(f"  {i}. {industry:40s} {std:.4f}")

print(f"  {output_file}")
```
